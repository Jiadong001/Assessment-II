### 比较 TransPHLA 和这四种模型的预测准确度

在independent_set2上

```python
import complete_models as cmodels

file_name = "data_csv/independent_set2.csv"

pre_NetMHCpan, accuracy_NetMHCpan = cmodels.use_NetMHCpan_10(file_name)
print("accuracy on NetMHCpan1.0 is {:.4f}".format(accuracy_NetMHCpan))

pre_LSTM, accuracy_LSTM = cmodels.use_LSTM(file_name)
print("accuracy on NetMHCpan1.0_LSTM is {:.4f}".format(accuracy_LSTM))

pre_BiLSTM, accuracy_BiLSTM = cmodels.use_BiLSTM(file_name)
print("accuracy on NetMHCpan1.0_BiLSTM is {:.4f}".format(accuracy_BiLSTM))

pre_AttBiLSTM, accuracy_AttBiLSTM = cmodels.use_AttBiLSTM(file_name)
print("accuracy on NetMHCpan1.0_AttBiLSTM is {:.4f}".format(accuracy_AttBiLSTM))

pre_TransPHLA, accuracy_TransPHLA = cmodels.use_TransPHLA(file_name)
print("accuracy on TransPHLA is {:.4f}".format(accuracy_TransPHLA))
```

```python
import matplotlib.pyplot as plt

accuracy = [accuracy_NetMHCpan, accuracy_LSTM, accuracy_BiLSTM, accuracy_AttBiLSTM, accuracy_TransPHLA]
name = ["NetMCpan1.0", "LSTM", "BiLSTM", "AttBiLSTM", "TransPHLA"]
color = ["green", "orange", "blue", "black", "red"]

bar_plot = plt.bar(name, accuracy, color=color, width=1, label='value')
plt.bar_label(bar_plot, label_type='edge')
plt.ylabel("accuracy")
plt.ylim(0, 1)
plt.title("prediction on independent_set2")

plt.show()
```
